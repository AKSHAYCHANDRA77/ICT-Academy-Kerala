```python
#Importing visualization libraries like matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot histogram
df.hist(bins=15, figsize=(10, 8), grid=False)
plt.show()
# Can understand the skewness of each column
```

```python
plt.figure(figsize=(15,12))
plt.scatter(df['Glucose'], df['Insulin'], c=df['Outcome'], cmap='coolwarm')
plt.title('Scatter Plot: Glucose vs Insulin')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.colorbar(label='Outcome')
plt.show()
# Scatter plot between 'Glucose' and 'Insulin', colored by 'Outcome'
```

```python
# Pair plot of selected features, color-coded by 'Outcome'
sns.pairplot(df[['Glucose', 'BMI', 'Insulin', 'Age', 'Outcome']], hue='Outcome', palette='coolwarm')
plt.show()
```

```python
# Heatmap for identify correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
