**Outlier detection and removal**

```python
# Box plot is used to identify ourlier in the column and using IQR method outliers are removed
sns.boxplot(x="ApplicantIncome",data=df)
```

```python
Q1 = df['ApplicantIncome'].quantile(0.25)
Q3 = df['ApplicantIncome'].quantile(0.75)
Q1, Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
df[(df.ApplicantIncome<lower_limit)|(df.ApplicantIncome>upper_limit)]
```

```python
# Removed the outliers
df=df[(df.ApplicantIncome>lower_limit)&(df.ApplicantIncome<upper_limit)]
df
```
